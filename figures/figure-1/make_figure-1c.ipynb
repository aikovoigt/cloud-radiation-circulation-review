# Maps of observed time-mean top-of-atmosphere cloud-radiative effects

This notebook reproduces Figure 1c.

Data: CERES EBAF edition 4.1, time-mean between 2007 and 2018

### Load libraries

```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
from cartopy.util import add_cyclic_point
```

For reference, print package versions to screen:

```python
print('xarrary:   ', xr.__version__)
print('numpy:     ', np.__version__)
import matplotlib; print('matplotlib:', matplotlib.__version__); del matplotlib
import cartopy; print('cartopy:   ', cartopy.__version__); del cartopy
```

### Load CERES data

```python
ds = xr.open_dataset('../../data/obs/CERES_EBAF_Ed4.1_200701-201812.timmean.cre_toa_sfc.nc').squeeze()
```

### Plot maps of TOA CRE

```python
clev=np.array([-90,-80,-70,-60,-50,-40,-30,-20,-10,10,20,30,40,50,60,70,80,90])
```

Shortwave

```python
plt.figure(figsize=(5.915, 3))
pdata, lons = add_cyclic_point(ds['toa_cre_sw_mon'].values, coord=ds.lon)
lats = ds.lat
ax = plt.axes(projection=ccrs.Robinson())
plt.contourf(lons, lats, np.clip(pdata,a_max=90,a_min=-90.0), clev,
             transform=ccrs.PlateCarree(), cmap='RdBu_r', extend='both')
ax.coastlines(resolution='110m', linewidth=0.75)
cb=plt.colorbar(aspect=25,fraction=0.03, format='% 1.0f')
cb.set_ticks([-90,-70,-50,-30,0,30,50,70,90]) # vertically oriented colorbar
cb.ax.tick_params(labelsize=8)
plt.title('Shortwave CRE at TOA', fontsize=12)
ax.text(1.05,-0.13,r'Wm$^{-2}$', fontsize=8,transform=ax.transAxes)
plt.savefig('figure-1c_toa_swcre_map.pdf')
```

Longwave

```python
plt.figure(figsize=(5.915, 3))
pdata, lons = add_cyclic_point(ds['toa_cre_lw_mon'].values, coord=ds.lon)
lats = ds.lat
ax = plt.axes(projection=ccrs.Robinson())
plt.contourf(lons, lats, pdata, clev,
             transform=ccrs.PlateCarree(),cmap='RdBu_r', extend='both')
ax.coastlines(resolution='110m')
cb=plt.colorbar(aspect=25,fraction=0.03, format='% 1.0f')
cb.set_ticks([-90,-70,-50,-30,0,30,50,70,90]) # vertically oriented colorbar
cb.ax.tick_params(labelsize=8)
plt.title('Longwave CRE at TOA', fontsize=12)
ax.text(1.05,-0.13,r'Wm$^{-2}$', fontsize=8,transform=ax.transAxes)
plt.savefig('figure-1c_toa_lwcre_map.pdf')
```
